```python
#9 Filter and select rows from the People_Dataset, where the “Last Name' column contains the name 'Duke', 'Gender' column contains the word Female and ‘Nalary’ should Pe less than 85000.


import pandas as pd

# Creating a sample dataset
data = {
    'First Name': ['John', 'Jane', 'Paul', 'Emily', 'Diana', 'Jessica'],
    'Last Name': ['Doe', 'Duke', 'Smith', 'Duke', 'Brown', 'Duke'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Female', 'Female'],
    'Salary': [95000, 82000, 67000, 88000, 45000, 78000]
}

# Convert the dictionary to a Pandas DataFrame
People_Dataset = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset:")
print(People_Dataset)
# Create an empty list to hold filtered rows
filtered_rows = []

# Iterate through each row in the DataFrame
for index, row in People_Dataset.iterrows():
    # Check the conditions for each row using an if statement
    if 'Duke' in row['Last Name'] and 'Female' in row['Gender'] and row['Salary'] < 85000:
        # If conditions are met, append the row to the filtered_rows list
        filtered_rows.append(row)

# Convert the filtered rows back into a DataFrame
filtered_data = pd.DataFrame(filtered_rows)

# Display the filtered data
print("\nFiltered Dataset")
print(filtered_data)

```

```python
#14 Create time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
#b) Label the x-axis as 'Date'.
#c) Set the title of the plot as 'Temperature and Humidity Over Time'.



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# a) Create a time-series dataset in a DataFrame with columns: 'Date', 'Temperature', and 'Humidity'
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')  # 30 days from Jan 1, 2023
temperature = np.random.randint(15, 35, size=30)  # Random temperatures between 15 and 35 degrees
humidity = np.random.randint(40, 90, size=30)  # Random humidity values between 40% and 90%

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plot Temperature and Humidity on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')  # Label x-axis as 'Date'
ax1.set_ylabel('Temperature (°C)', color='tab:red')  # Label left y-axis for Temperature
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()  # Create a twin axes sharing the same x-axis
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')  # Label right y-axis for Humidity
ax2.tick_params(axis='y', labelcolor='tab:blue')

# c) Set the title of the plot as 'Temperature and Humidity Over Time'
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()

```

```python
#Que 15--> Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
 #a) Plot a histogram of the data with 30 bins.
 #b) Overlay a line plot representing the normal distribution's probability density function (PDF).
 #c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 #d) Set the title of the plot as 'Histogram with PDF Overlay'.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # loc is mean, scale is standard deviation

# Step 2: Plot a histogram of the data with 30 bins
plt.figure(figsize=(8, 6))
count, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Step 3: Overlay a line plot for the normal distribution's PDF
xmin, xmax = plt.xlim()  # Get the range for x-axis from the histogram
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, np.mean(data), np.std(data))  # PDF of the normal distribution
plt.plot(x, pdf, 'r-', label='PDF')

# Step 4: Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Step 6: Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Que16 -> Set the title of the plot as 'Histogram with PDF Overlay'.

#Ans-->
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generating 1000 samples from a normal distribution
data = np.random.randn(1000)

# Plot the histogram
count, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Overlay the PDF (Probability Density Function)
mu, sigma = norm.fit(data)  # Fit a normal distribution to the data
pdf = norm.pdf(bins, mu, sigma)  # Get the PDF values
plt.plot(bins, pdf, 'r--', linewidth=2, label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Show the plot
plt.show()

```

```python
#Que17--> Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
#Ans -->

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate two random arrays
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame to store the values
df = pd.DataFrame({'x': x, 'y': y})

# Define a function to classify points into quadrants
def classify_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant 1'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant 2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

# Apply the function to classify each point
df['Quadrant'] = df.apply(classify_quadrant, axis=1)

# Create the scatter plot using Seaborn
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1', s=100)

# Label the axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Add the legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()

```

```python
#Que19 --> Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
#Bar Chart'

#Ans->

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
import numpy as np

output_notebook()


categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a data source for Bokeh
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a Bokeh figure (corrected attributes)
p = figure(x_range=categories, height=400, width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars with color mapping based on values
p.vbar(x='categories', top='values', width=0.6, source=source,
       fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tool to display the exact value
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the plot
show(p)


```

```python
#Que20 -> Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.


import plotly.graph_objs as go
import numpy as np
import plotly.offline as pyo

# Generate random data for the line plot
x = np.arange(1, 101)
y = np.random.randint(10, 50, size=100)
line_plot = go.Scatter(x=x, y=y, mode='lines', name='Line Plot')
layout = go.Layout(
    title='Simple Line Plot',  # Title of the plot
    xaxis=dict(title='X-axis'),  # Label for the x-axis
    yaxis=dict(title='Y-axis')   # Label for the y-axis
)

# Create a figure object with data and layout
fig = go.Figure(data=[line_plot], layout=layout)

# Display the plot
pyo.plot(fig)
```
